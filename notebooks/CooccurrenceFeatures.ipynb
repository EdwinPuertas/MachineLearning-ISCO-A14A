```python
data = pd.read_csv(DIR_DATA + 'tass2018_es_train.csv', sep=',')
data
```

```python
data = [tp.transformer(row, stopwords=True) for row in tqdm(data['content'].tolist())]
len(data)
```

```python
df = pd.DataFrame(Xc.toarray(),
                  index=['content '+str(i) for i in range(1, 1+len(coo_vector.vocabulary_))],
                  columns=coo_vector.get_feature_names())
```
